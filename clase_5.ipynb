Inmigration pattern to Canada

```python
path = '/content/drive/MyDrive/Canada.xlsx'
```

```python
import pandas as pd
```

```python
df = pd.read_excel(path,
                   sheet_name = 'Canada by Citizenship',
                   skiprows = range(20),
                   skipfooter = 2
)
df
```

#Description of the data set

```python
df.columns
```

```python
df.head(2)
```

```python
to_be_dropped = ['Type','Coverage','AREA','REG','DEV','DevName']
```

```python
type(to_be_dropped)
```

```python
df.drop(to_be_dropped,
        axis = 1,
        inplace = True)
df.head(2)
```

```python
new_names = {'OdName':'Country',
             'AreaName':'Continent',
             'RegName':'Region'
}
```

```python
df.rename(columns = new_names,
          inplace = True
)
df.head(2)
```

#Description of the variables

#How is the immigration to Canada from Japan during 1980 to 2013??

```python
df.head(2)
```

```python
df['Total'] = df.iloc[:,3:].sum(axis=1)
df.head(2)
```

```python
df.set_index('Country', inplace = True)
df.head(2)
```

```python
df.loc['Japan',1980:2013].plot(kind='line')
```

#Conclusion:
There was one decrease in 1985, then gradual increase from 1985 to 2000...
